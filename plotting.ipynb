1. Create a Time Series Plot. Using the data contained in the file time_and_temp.txt plot the data in the file and use the text information in the file to construct a title, x-axis label and y-axis label. Also construct a plot code to draw a dashed line with ‘o’ as data markers and the color set to green.  Notice the plot shows a time history of the data that was collected.


```python
import matplotlib.pyplot as plt


time = []
temp = []
#loop thru each line in data set
with open("time_and_temp.txt", "r") as f:
    next(f)
    for line in f:
        stri_parts = line.strip()
        parts = stri_parts.split()
        if len(parts) >= 2:
            tim = float(parts[0])
            tem = float(parts[1])
            time.append(tim)
            temp.append(tem)

    

xmin = min(time)
xmax = max(time)
ymin = min(temp)
ymax = max(temp)

plt.plot(time, temp, color='green', linestyle='--', marker='o')
plt.title("Time and Temperature")
plt.xlabel(time[0])
plt.ylabel(temp[0])

plt.axis([xmin, xmax, ymin, ymax])
plt.show()
```

```python
import matplotlib.pyplot as plt

temperature = []
pressure = []


with open('temp_and_pressure.txt', 'r') as f:
    next(f)
    for line in f:
        parts = line.split()
        if len(parts) == 2:
            temp = (parts[0])
            pres = (parts[1])
            temperature.append(float(temp))
            pressure.append(float(pres))

            

avg_temp = sum(temperature)/ len(temperature)
avg_pres = sum(pressure)/ len(pressure)

#limits
xmin = min(temperature)
xmax = max(temperature)
ymin = min(pressure)
ymax = max(pressure)

#averafge lines
plt.plot([xmin, xmax],[avg_pres, avg_pres], "--k")
plt.plot([avg_temp, avg_temp], [ymin, ymax], "--k")

plt.scatter(temperature, pressure, color='red', marker='o', label='data points')
plt.plot(avg_temp, avg_pres, "o")
plt.text(avg_temp + 0.5, avg_pres + 0.5, 'data centroid', fontsize=10, color='blue')

plt.title("Pressure vs. Temperature")
plt.xlabel("Temperature (C)")
plt.ylabel("Pressure (lbs/in*in)")
plt.plot(temperature, pressure, "o", color = "red")
plt.show()
```

```python
import matplotlib.pyplot as plt


time = []
temperature = []
pressure = []
volume = []


with open('time_temp_pres_vol.txt', 'r') as f:
    next(f)  # skip header, 
    for line in f:
        partstrip = line.strip()
        parts = partstrip.split()
        if len(parts) == 4:
            t = float(parts[0])
            temp = float(parts[1])
            pres = float(parts[2])
            vol = float(parts[3])
            time.append(t)
            temperature.append(temp)
            pressure.append(pres)
            volume.append(vol)


avg_temp = sum(temperature)/ len(temperature)
avg_pres = sum(pressure)/ len(pressure)
avg_vol = sum(volume)/ len(volume)
         
#subplot 1
plt.subplot(3, 1, 1)
plt.plot(time, temperature, linestyle = "-", color = "red")
plt.title("Temp vs. time")
plt.ylabel("Temp (C)")
plt.axis([min(time), max(time), min(temperature), max(temperature)])
plt.plot([min(time), max(time)], [avg_temp, avg_temp])
#subplot 2
plt.subplot(3, 1, 2)
plt.plot(time, pressure, linestyle = "-", color='green')
plt.title("Pressure vs. time")
plt.ylabel("Pressure (lbs/in)")
plt.axis([min(time), max(time), min(pressure), max(pressure)])
plt.plot([min(time), max(time)], [avg_pres, avg_pres], color = "blue")
#3 vol vs. time
plt.subplot(3, 1, 3)
plt.plot(time, volume, linestyle = "-", color = "magenta")
plt.title("Volume vs. time")
plt.xlabel("TIme (secs)")
plt.ylabel("Volume ")
plt.axis([min(time), max(time), min(volume), max(volume)])
plt.plot([min(time), max(time)], [avg_vol, avg_vol], color = "blue")
plt.show()


```

```python
import matplotlib.pyplot as plt
names = ["Bill", "Sue", "Mary", "Jane", "Dave", "John"]
sales_counts = [0, 0, 0,0,0, 0]
sales_amounts = [0,0,0,0,0, 0]

with open("sales_data.txt", "r") as f:
    next(f)
    for line in f:
        stripparts = line.strip()
        parts = stripparts.split()
        if len(parts) == 2:
            name = parts[0]
            amount = (float(parts[1]))
            if name in names:
                nameindex = names.index(name)
                sales_counts[nameindex] += 1
                sales_amounts[nameindex] += amount
exp_for_counts = [0] * len(sales_counts)
exp_for_amount = [0] * len(sales_amounts)
exp_for_counts[sales_counts.index(max(sales_counts))] = 0.1
exp_for_amount[sales_amounts.index(max(sales_amounts))] = 0.1
#sales count per person
plt.subplot(1, 2, 1)
plt.title("Sales count per salesperson")
plt.pie(sales_counts, explode = exp_for_counts, labels = names)
#sales amount per perso
plt.subplot(1, 2, 2)
plt.title("Sales amount per salesperson ($)")
plt.pie(sales_amounts, explode = exp_for_amount, labels = names)

plt.show()
```
